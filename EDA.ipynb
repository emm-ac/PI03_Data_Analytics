Tablas correspondientes al segundo KPI

```python
#Importar CSVs y revisar datos.
data2_internet = pd.read_csv('./CSV_sin revisar/2_internet.csv')
data2_telefonia = pd.read_csv('./CSV_sin revisar/2_telefonia.csv')
data2_television = pd.read_csv('./CSV_sin revisar/2_television.csv')
```

```python
#Análisis exploratorio para conocer la estructura de los datos y ver si hay errores.
data2_internet.info()
data2_internet.value_counts()
data2_internet.isna().sum()
```

```python
#Reordenar los datos y actualizar índices.
data2_internet = data2_internet.sort_values(['Año','Trimestre','Provincia'])
data2_internet.reset_index(drop=True, inplace=True)
data2_internet.head()
```

```python
#Análisis exploratorio para conocer la estructura de los datos y ver si hay errores.
data2_telefonia.info()
data2_telefonia.value_counts()
data2_telefonia.isna().sum()
```

```python
#Reordenar los datos y actualizar índices.
data2_telefonia = data2_telefonia.sort_values(['Año','Trimestre','Provincia'])
data2_telefonia.reset_index(drop=True, inplace=True)
data2_telefonia.head()
```

```python
#Análisis exploratorio para conocer la estructura de los datos y ver si hay errores.
data2_television.info()
data2_television.value_counts()
data2_television.isna().sum()
```

```python
#Reordenar los datos y actualizar índices.
data2_television = data2_television.sort_values(['Año','Trimestre','Provincia'])
data2_television.reset_index(drop=True, inplace=True)
data2_television.head()
```

```python
#Armar una tabla única con todos los datos.
data2_telefonia = data2_telefonia[['Accesos telefonía fija por cada 100 hab','Accesos telefonía fija por cada 100 hogares']]
data2_television = data2_television[['Accesos TV por suscripción por cada 100 hab','Accesos TV por suscripción por cada 100 hogares']]
total = [data2_internet,data2_telefonia,data2_television]
data2 = pd.concat(total, axis=1)
data2.head()
```

```python
#Exportar datos listos para trabajar.
data2.to_csv('./CSV_ok/KPI2.csv')
```

Tablas correspondientes al cuarto KPI

```python
#Importar CSVs y revisar datos.
data4_tecnologia = pd.read_csv('./CSV_sin revisar/4_internet_tecnologia.csv')
data4_velocidad = pd.read_csv('./CSV_sin revisar/4_internet_velocidad.csv')
```

```python
#Análisis exploratorio para conocer la estructura de los datos y ver si hay errores.
data4_tecnologia.info()
data4_tecnologia.value_counts()
data4_tecnologia.isna().sum()
```

```python
#Reordenar los datos y actualizar índices.
data4_tecnologia.dropna(inplace=True)
data4_tecnologia['Año'] = data4_tecnologia['Año'].apply(lambda x: x.replace(' *',''))
data4_tecnologia['Trimestre'] = data4_tecnologia['Trimestre'].apply(lambda x: x.replace(' *',''))
data4_tecnologia.rename(columns={'Otros':'Otros (tecnología)', 'Total':'Total (tecnología)'}, inplace=True)
data4_tecnologia = data4_tecnologia.sort_values(['Año','Trimestre','Provincia'])
data4_tecnologia.reset_index(drop=True, inplace=True)
data4_tecnologia.head()
```

```python
#Verificar que no haya valores nulos.
data4_tecnologia.info()
data4_tecnologia.value_counts()
data4_tecnologia.isna().sum()
```

```python
#Análisis exploratorio para conocer la estructura de los datos y ver si hay errores.
data4_velocidad.info()
data4_velocidad.value_counts()
data4_velocidad.isna().sum()
```

```python
#Reordenar los datos y actualizar índices.
data4_velocidad = data4_velocidad.sort_values(['Año','Trimestre','Provincia'])
data4_velocidad.reset_index(drop=True, inplace=True)
data4_velocidad.rename(columns={'OTROS':'Otros (velocidad)', 'Total':'Total (velocidad)'}, inplace=True)
data4_velocidad.head()
```

```python
#Armar una tabla única con todos los datos.
data4_velocidad.drop(['Año','Trimestre','Provincia'],axis=1,inplace=True)
total4 = [data4_tecnologia,data4_velocidad]
data4 = pd.concat(total4, axis=1)
data4.head()
```

```python
#Exportar datos listos para trabajar.
data4.to_csv('./CSV_ok/KPI4.csv')
```
